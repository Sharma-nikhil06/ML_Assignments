```python
# Question 1
import numpy as np
import pandas as pd
import re
```

```python
a=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv ')
b=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
```

```python
c=pd.concat([a,b])
c.head()
```

```python
# Question 2
c=c.dropna(subset=['Survived'])
c.head()
```

```python
c.info
```

```python
c['Has_Cabin']=c['Cabin'].notnull()
```

```python
c.info
```

```python
def split_it(year):
    return str(re.findall("(Mr|Mrs|Master|Miss)",year))
c["Title"]=c['Name'].apply(split_it)
```

```python
c['Title'].value_counts()
```

```python
c=c.fillna(c.mean())
```

```python
c['Categorical_fare']=pd.cut(c['Fare'], 3,labels=[ "Economy" ,"Premium-Economy","Business"])
```

```python
c['Categorical_age']=pd.cut(c['Age'], [0,13,19,100], labels=["Child", "Teen", "Adult"])
```

```python
c=c.drop(['Cabin','Name','PassengerId','Ticket','Age','Fare','SibSp','Parch'],axis=1)
```

```python
c=pd.get_dummies(c)
```

```python
x=c.drop(["Survived"],axis=1)
y=c["Survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
score=classifier.score(X_test,y_test)
print(score)
```
